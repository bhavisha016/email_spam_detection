```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv(r"D:\csv files\email.csv");
df.head()
```

```python
df['Spam']=df['Category'].apply(lambda x: 1 if x=="spam" else 0)#y part has been converting to numbers as model understands numbers not text
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.Message,df['Spam'],train_size=0.8)
```

```python
x_train[:4]
```

```python
v=CountVectorizer()##bow counts freq of words
x_train_cv=v.fit_transform(x_train)##csr or compressed sparse row only stores The positions of non-zero values
print(x_train_cv.shape)#4458 messages 7728 unique words
```

```python
x_train_cv[0].toarray()##shows count for 1st meassage but in csr only non zero values were stored to save memory
```

```python
x_train[0]##first message and above is the freq count of thats first message
```

```python
vocab = v.get_feature_names_out()##all vocabs 
# Get the 1st row as an array (dense)
row_vector = x_train_cv[0].toarray()[0]

# Get the words with non-zero counts (i.e., words used in message 1)
words_in_row = [vocab[i] for i, count in enumerate(row_vector) if count > 0]

print("ðŸŒŸ Words from 1st message found in vocabulary:")
print(words_in_row)
```

```python
model=MultinomialNB()##train model using naive bayes
model.fit(x_train_cv,y_train)
```

```python
x_test_cv=v.transform(x_test)
```

```python
y_predicted = model.predict(x_test_cv)
print(classification_report(y_test,y_predicted))
```
